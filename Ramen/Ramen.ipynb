```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
train = pd.read_csv('ramen-ratings.csv')
```

```python
train.head()
```

```python
train.drop('TopTen',axis=1,inplace=True)
train.drop('Brand',axis=1,inplace=True)
train.drop('Variety',axis=1,inplace=True)
```

```python
train.head()
```

```python
train.groupby('Style')['Stars'].count()
```

```python
def fillStyle(style):
    style = style[0]
    print(style)
    if style == "Cup":
        return 0
    elif style == "Bar":
        return 1
    elif style == "Bowl":
        return 2
    elif style == "Box":
        return 3
    elif style == "Can":
        return 4
    elif style == "Pack":
        return 5
    else:
        return 6
```

```python
train['Style']=train[['Style']].apply(fillStyle,axis=1)
```

```python
train.head()
```

```python
#train.groupby('Country')['Stars'].count()
```

```python
def fillCountry(country):
    country = country[0]
    if country=='Japan':
        return 0
    elif country=='USA':
        return 1
    elif country=='South Korea':
        return 2
    elif country=='Taiwan':
        return 3
    elif country=='Thailand':
        return 4
    elif country=='China':
        return 5
    else:
        return 6
train['Country']=train[['Country']].apply(fillCountry,axis=1)
```

```python
train.head()
```

```python
def fillStars(stars):
    stars = stars[0]
    if  "3" in stars:
        return 1
    if  "2" in stars:
        return 0
    if  "1" in stars:
        return 0
    else:
        return 1
```

```python
train['Stars']=train[['Stars']].apply(fillStars,axis=1)
```

```python
train.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
dummieStars = pd.get_dummies(train['Stars'],drop_first=False)
```

```python
X_train, X_test, y_train, y_test = train_test_split(train.drop('Stars',axis=1),
                                                    train['Stars'],test_size=0.99,
                                                    random_state=101)
                                                    
```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
predictions = logmodel.predict(X_test)
print(logmodel.score(X_train,y_train))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```
